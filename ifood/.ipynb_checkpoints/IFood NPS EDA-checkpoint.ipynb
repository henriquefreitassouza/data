# Quais alavancas vão promover a satisfação dos restaurantes com a plataforma IFood?

Nos últimos meses temos notado um NPS abaixo de nossas expectativas nas avaliações feitas pelos restaurantes. Precisamos entender as causas desse KPI baixo e possíveis alavancas para reversão do cenário.

**Hipóteses**

- A insatisfação dos clientes dos restaurantes aumentou, gerando também insatisfação dos restaurantes;
- Clientes estão comprando menos na plataforma, gerando insatisfação dos restaurantes;
- Os tempos de entrega dos parceiros IFood não estão agradando aos restaurantes.

## Importação das ferramentas

Serão utilizadas as bibliotecas Numpy e Pandas para conhecer os dados, e a PyPlot, do pacote Matplotlib, para visualizá-los.

## Gerar visualizações que explicam a dispersão do dataset

```python
df['nps_grade'].plot(kind='hist', bins=10)
plt.xlabel('nps_grade')
```

```python
df['months_at_ifood'].plot(kind='hist', bins=60)
plt.xlabel('months_at_ifood')
```

```python
df['months_at_ifood_int'].plot(kind='hist', bins=60)
```

A maioria dos restaurantes avaliados possuem menos de 10 meses na plataforma.

```python
df[df['months_at_ifood'] >= 0][['nps_grade', 'months_at_ifood']].plot(kind='scatter', x='nps_grade', y='months_at_ifood')
```

```python
# Visualizando a diferença das notas por tempo médio de cadastro
df[(df['months_at_ifood'] < 9) & (df['months_at_ifood'] >= 0)][['nps_grade', 'months_at_ifood']].plot(kind='scatter', x='nps_grade', y='months_at_ifood')
```

Enquanto o NPS médio é puxado para cima por avaliações acima de 5, o tempo médio na plataforma é puxado para baixo pela quantidade de novos cadastros. A sobreposição das duas variáveis indica que tempo na plataforma não é um fator para a obtenção de notas altas no NPS, mas restaurantes que estão na plataforma há mais de 60 meses costumam receber mais avaliações com notas 9 e 10.

```python
df['date_month'] = df.index.month
```

```python
new_restaurants = df[df['months_at_ifood'] < 9].groupby('date_month').agg({'restaurant_id': 'count', 'nps_grade': 'mean', 'months_at_ifood': 'mean'})
new_restaurants['nps_diff'] = new_restaurants['nps_grade'].diff()
new_restaurants
```

```python
old_restaurants = df[df['months_at_ifood'] > 59].groupby('date_month').agg({'restaurant_id': 'count', 'nps_grade': 'mean', 'months_at_ifood': 'mean'})
old_restaurants['nps_diff'] = old_restaurants['nps_grade'].diff()
old_restaurants
```

```python
between_new_and_old_restaurants = df[(df['months_at_ifood'] > 8) & (df['months_at_ifood'] <= 59)].groupby('date_month').agg({'restaurant_id': 'count', 'nps_grade': 'mean', 'months_at_ifood': 'mean'})
between_new_and_old_restaurants['nps_diff'] = between_new_and_old_restaurants['nps_grade'].diff()
between_new_and_old_restaurants
```

```python
df.groupby('date_month').mean()['nps_grade'].plot()
plt.ylabel('nps_grade')
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(new_restaurants['nps_grade'], label='New restaurants (< 9 months)')
ax.plot(between_new_and_old_restaurants['nps_grade'], label='Between old and new restaurants ( >= 9 months and <= 59 months)')
ax.plot(old_restaurants['nps_grade'], label='Old restaurants (> 59 months)')
ax.legend()
plt.xlabel('date_month')
plt.ylabel('nps_grade')
```

O NPS vem caíndo mês a mês quando considerado o quadro geral de avaliações. Ao separar por faixas, no entanto, é possível notar que a oscilação é mais acentuada em restaurantes antigos. Restaurantes com menos de 60 meses de cadastro começaram a perder nota enquanto restaurantes com mais de 60 meses de cadastro tiveram melhores avaliações. O movimento contrário acontece em Julho, quando novos restaurantes recuperam parte da nota e antigos caem quase 1,5 pontos na nota. De Agosto em diante os restaurantes se aproximam nas notas e todos apresentam tendência de queda no final do período analisado, com destaque para restaurantes antigos.
